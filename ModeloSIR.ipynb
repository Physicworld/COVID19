```
import scipy.integrate
import numpy as np
import matplotlib.pyplot as plt
```

```
def SIR_model(y, t, beta, gamma):
  S, I, R = y

  ds_dt = -beta*S*I
  di_dt = beta*S*I - gamma*I
  dr_dt = gamma*I

  return([ds_dt,di_dt,dr_dt])
```

```
# Condiciones Iniciales

N = 10
P0 =763

beta = 0.0022
gamma = 0.4477

S0 = 763
I0 = 0.1
R0 = 0

t = np.linspace(0,14,10000)

solucion = scipy.integrate.odeint(SIR_model,[S0,I0,R0], t, args=(beta, gamma))
solucion = np.array(solucion)
```

```
plt.figure(figsize=[8,6])
plt.plot(t, solucion[:,0], label="S(t)")
plt.plot(t, solucion[:,1], label="I(t)")
plt.plot(t, solucion[:,2], label="R(t)")
plt.grid()
plt.legend()
plt.xlabel("Dias transcurridos")
plt.ylabel("Proporciones")
plt.title("SIR MODEL")
plt.show()
```
